```python
class Solution(object):
    def moveZeroes(self, nums):
        for i in range(len(nums)):
            if nums[i]==0:
                nums[i],nums[-1]=nums[-1],nums[i]
        return nums
```

```python
class Solution(object):
    def moveZeroes(self, nums):
        l=len(nums)
        m=0
        for i in range(l):
            if nums[i]!=0:
              if i!=m:
                nums[i],nums[m]=nums[m],nums[i]
                m+=1
        return nums
```
